```python
import requests as rq

from bs4 import BeautifulSoup
```

```python
url = 'http://forum.yurclub.ru/'
```

```python
page = rq.get(url)
```

```python
page 
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for tag in soup.find_all('span', itemprop = 'name'):
    print(tag.text)
```
